```python
import os, shutil
import sys
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import MaxPool2D, Conv2D, Dense, Flatten
```

#### Arranging your data and directory structure
___
##### Note: do this only if you are working in your local system

```python
test = 'evaluation'
train = 'train'
val = 'validation'
base_dir = 'dataset'
```

```python
def make_directory_structure(base_dir, train, val, test):
    os.mkdir(base_dir)
    counters = [4999, 999, 999]
    for c , folder in enumerate([train, val, test]):
        folder_path = os.path.join(base_dir, folder)
        os.mkdir(folder_path)
        for cat in ['cat', 'dog']:
            os.mkdir(os.path.join(base_dir, folder, cat))
            for i, file in enumerate(os.listdir(os.path.join('data/cats_and_dogs',folder, cat))):
                if i > counters[c]:
#                     print(i,c)
                    break
                else:
                    src = os.path.join('data/cats_and_dogs',folder, cat, file)
                    dst = os.path.join(base_dir, folder, cat, file)
#                     print(src, dst)
                    shutil.copyfile(src, dst)
```

```python
make_directory_structure(base_dir, train, val, test)
```

```python
def check_dataset(base_dir, train, val, test):
    print(f'#train/cat: {len(os.listdir(os.path.join(base_dir,train,"cat")))}')
    print(f'#train/cat: {len(os.listdir(os.path.join(base_dir,train,"cat")))}')
    print(f'#validation/cat: {len(os.listdir(os.path.join(base_dir,val,"cat")))}')
    print(f'#validation/cat: {len(os.listdir(os.path.join(base_dir,val,"cat")))}')
    print(f'#test/cat: {len(os.listdir(os.path.join(base_dir,test,"cat")))}')
    print(f'#test/cat: {len(os.listdir(os.path.join(base_dir,test,"cat")))}')
#     print(f'#validation/cat: {len(os.listdir(os.path.join(base_dir,val,"cat")))}')
```

```python
check_dataset(base_dir, train, val, test)
```

#### Building the model

```python
model = Sequential()
model.add(Conv2D(32, (3,3), activation='relu', input_shape=(150,150,3)))
model.add(MaxPool2D(2,2))
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPool2D(2,2))
model.add(Conv2D(128, (3,3), activation='relu'))
model.add(MaxPool2D(2,2))
model.add(Conv2D(128, (3,3), activation='relu'))
model.add(MaxPool2D(2,2))
model.add(Flatten())
## fully connected layers
model.add(Dense(512, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile('rmsprop',
              'binary_crossentropy',
              metrics=['accuracy']
             )
```

```python
model.summary()
```

#### Finally Trainning the model

```python
history = model.fit_generator(train_generator, epochs=5, steps_per_epoch=60, validation_data=validation_generator, validation_steps=30)
```

### Using data augmentation

If we do the same for more number of epochs we will be in a better position to decide whether *to look for a better model* or *to collect more data for trainning*. The general Idea is explained as followed:

![](images/data_agumentation.png)

Now we have a clear idea that our model is overfitting (first case, we need more data !). Overfitting is caused by having too few samples to learn from. Given infinite data, your model would be wxposed to every possible aspect of the data distribution at hand. In such situations we can use data augmentation.

Data Augmentation takes the approach of generating more training data from existing training samples, by augmenting the samples via a number of random transformations that yield believable-loking images, The goal is that at training time, your model will never see the exact same picture twice. This helps expose the model to more aspects of the data and generalize better.

In keras, its super easy!!

```python
datagen = ImageDataGenerator( rotation_range=40,
                              width_shift_range=0.2,
                              height_shift_range=0.2,
                              shear_range=0.2,
                              zoom_range=0.2,
                              horizontal_flip=True,
                              fill_mode='nearest')
```

```python
from keras.preprocessing import image
```

```python
frame_name = os.listdir('dataset/evaluation/cat/')[10]
img = image.load_img(os.path.join('dataset/evaluation/cat/', frame_name), target_size=(150,150))
X = image.img_to_array(img) # converting img to numpy array
X = X.reshape((1,)+X.shape) # reshape to (1,150,150,3)
```

```python
datagen.fit(X)
```

```python
i = 0
plt.figure(figsize=(10,8))
for batch in datagen.flow(X, batch_size=1):
#     print(batch.shape)
    plt.subplot(221+i)
    plt.imshow(image.array_to_img(batch[0]))
    
    i += 1
    if i%4 == 0:
        break
```

For more on data augmentaion, refer [this link](https://machinelearningmastery.com/image-augmentation-deep-learning-keras/)

```python
train_datagen = ImageDataGenerator(rescale=1./255,
                                   rotation_range= 40,
                                  width_shift_range = 0.2,
                                  height_shift_range = 0.2,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)

train_generator = train_datagen.flow_from_directory(os.path.join(base_dir,train),
                                                     target_size=(150,150), 
                                                     batch_size=32,
                                                     class_mode='binary')

validation_datagen = ImageDataGenerator(rescale=1./255,
                                   rotation_range= 40,
                                  width_shift_range = 0.2,
                                  height_shift_range = 0.2,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)

validation_generator = validation_datagen.flow_from_directory(os.path.join(base_dir,val),
                                                               target_size=(150,150),
                                                               batch_size=32,
                                                               class_mode= 'binary')
```
